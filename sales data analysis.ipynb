```python
# How many sales per month do we have?
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
data_month_oreder_freq = data["Month"].value_counts()
data_month_oreder_freq.index = pd.Categorical(data_month_oreder_freq.index,categories = month_order,ordered=  True ) #based on frequancy
data_month_oreder_freq.sort_index()
```

```python
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]

data_sum = data.groupby("Month")["Order_Quantity"].sum()  #based on orderd quantity
data_sum.index=pd.Categorical(data_sum.index,categories=month_order,ordered=True)
monthly_order_sum = data_sum.sort_index(ascending=False)
monthly_order_sum.head(1)
```

```python
#  show a bar plot
monthly_order_sum.plot(kind="bar",figsize=(14,6))
```

```python
# How many orders were made in Canada or France?
data.groupby("Country")["Order_Quantity"].sum().loc[['France','Canada']]
```

```python
# How many Bike Racks orders were made from Canada?
# approach 1 by groupby
data_cou_pro = data.groupby(["Country","Sub_Category"])["Order_Quantity"].sum().reset_index()
data_cou_pro.loc[(data_cou_pro["Country"]=="Canada")&(data_cou_pro["Sub_Category"]=="Bike Racks")].head(1)
```

```python
filtered_data = data[(data["Country"]=="Canada")&(data["Sub_Category"]=="Bike Racks")]
filtered_data.value_counts()
```

```python
# How many orders were made in each region (state) of France?
filtered_france_data = data[(data["Country"]=="France")]
filtered_france_state_data = filtered_france_data.groupby("State")["Order_Quantity"].sum().sort_values(ascending= False).reset_index()
filtered_france_state_data
```

```python
# show a bar plot with the results:
filtered_france_state_data.plot(kind="bar",x = "State")
plt.xlabel("State")
plt.ylabel("Order Quantity")
plt.title("France State wise sales data")
plt.show()
```
